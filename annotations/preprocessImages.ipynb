```python
from PIL import Image
import os
import numpy as np
import cv2
```

```python
with Image.open('input_image.tif') as img:
    # Convert the image to RGB mode if its in a different mode like RGBA
    img = img.convert('RGB')
    img = img.quantize(colors=256) #convert to 8-bit
    # Save as JPEG
    img.save('output_image.jpg', 'JPEG')
```

```python
for folder in genotype_folders:
    print(folder)
    for x in os.listdir('data/' + folder):
        if x.lower().endswith('.tif'):
            file_path = os.path.join('data/' + folder, x)
            #with Image.open(file_path) as img:
                #display(img)
                #print(img.mode)
                #img2 = img.point(lambda i: i * (1 / 256)).convert('L') # Convert 16-bit grayscale to 8-bit grayscale
                #display(img2)
                #print(img2.mode)
                #img3 = img2.convert('RGB')
                #display(img3)
                #print(img3.mode)
                # Define new filename with .jpg extension
                #new_filename = os.path.splitext(x)[0] + '.jpg'
                #new_file_path = os.path.join('data/' + folder, new_filename)
                #img3.save(new_file_path, 'JPEG')
            with Image.open(file_path) as img:
                # Convert PIL image to numpy array
                img_np = np.array(img)
                #display(img_np)

                # Normalize the image to 8-bit
                img_np = cv2.normalize(img_np, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
                #display(img_np)

                # Convert grayscale to RGB
                img_rgb = cv2.cvtColor(img_np, cv2.COLOR_GRAY2RGB)
                #display(img_rgb)

                # Convert numpy array back to PIL image
                img_pil = Image.fromarray(img_rgb)
                #display(img_pil)
                new_filename = os.path.splitext(x)[0] + '.jpg'
                new_file_path = os.path.join('data/' + folder, new_filename)
                img_pil.save(new_file_path, 'JPEG')
```
